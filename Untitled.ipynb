```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import zipfile

plt.style.use('fivethirtyeight') # Use plt.style.available to see more styles
sns.set()
sns.set_context("talk")
%matplotlib inline
```

```python
# load both singal.csv and tresspasser.csv files
df_signal = pd.read_csv("data/signal.csv", index_col=0)
df_tresspasser02 = pd.read_csv("data/trespasser/221002/Midland.csv", index_col=0)
display(df_signal.head())
display(df_tresspasser02.head())
```

```python
# compare the datat in "10/2/2022"
# 第一种情况, 当在火车栅栏放下来后,有没有tres的数据

df_signal_1002 = df_signal[df_signal['start_time'].str.match('^10/2/2022.*M$')== True]
display(df_signal_1002.head())
df_tress_1002 = df_tresspasser02[df_tresspasser02['start_time'].str.match('^10/2/2022.*M$')== True]
#df_tress_1002 = df_signal.loc[df_signal['validation'] == some_value]
display(df_tress_1002.head())
```

```python
# Change the string to datetime
df_signal_1002['start_time'] = pd.to_datetime(df_signal_1002['start_time'])
df_signal_1002['end_time'] = pd.to_datetime(df_signal_1002['end_time'])
```

```python
df_signal_1002.info()
df_signal_1002.head()
```

```python
# Calculate "duration"
df_signal_1002['duration'] = (df_signal_1002['end_time']-df_signal_1002['start_time'])
display(df_signal_1002.head())
```

```python
df_signal_1002.plot(x='start_time', y='duration', style=".")
```

```python
# try to find "signal and tress overlap" datat

df_signal_1002['star_time_short']=df_signal_1002 ['start_time'].str.extract(r'(^.*:.*:)')
display(df_signal_1002.head())
df_tress_1002['star_time_short']=df_tress_1002 ['start_time'].str.extract(r'(^.*:.*:)')
display(df_tress_1002.head())
```

```python
df_signal_1002['new'] = np.where((df_signal_1002['star_time_short'] == df_tress_1002['star_time_short']), 'Overlap', np.nan)
display(df_signal_1002)
```
